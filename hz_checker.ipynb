```python
import math

def q_cur_n(f1, f2, n):
    return math.floor((n * f2) / f1)

def q_use_n(q_cur): 
    return max(q_cur-1, 0)

def n_obs_n(q_use, f1, f2):
    return math.ceil((q_use * f1) / f2)


def hz_paser(n:int = None, f1:float = 15, f2:float = 10):
    q_cur = q_cur_n(f1, f2, n)
    q_use = q_use_n(q_cur)
    n_obs = n_obs_n(q_use, f1, f2)
    return n_obs

for n in range(301):
    n_obs = hz_paser(n)
    print('* n :', n+1, '=> latent_n :', n_obs)
```
